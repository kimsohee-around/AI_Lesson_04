### 문자열 불변객체
- 문자열은 불변(immutable) 객체 입니다.

```python
a="Hello"
id(a)
```

```python
a = a+', python!'
id(a)   # 문자열 연산 결과를 다른 메모리 위치에 저장합니다.
```

```python
text = "Hello"
id(text)
```

```python
text = "Python"
id(text)
```

### 문자열 iterable 객체
- iterable 객체는 반복 및 순차적으로 접근할 수 있는 객체 입니다.
- 인덱스 [ ] 기호 사용 가능(파이썬 특징)

```python
my_string ="Hello"
my_string[2]
```

```python
my_string[-4]   # 문자열의 마지막 인덱스값 리턴
# 인덱스값 -1이면 맨 뒤. -2, -3 은 맨뒤부터 인덱스 값 지정
```

```python
# my_string[0]='h'   # 오류: 값 변경은 불가
```

```python
new_string ='h' + my_string[1:]
new_string
```

### 문자열 함수와 연산

```python
# 문자열 길이
text = "Hello, Python!"
len(text)
```

```python
# 문자열 반복
apple = "사과 " * 10
apple
```

```python
# 문자열 인덱싱과 슬라이싱
#       0 1 2 34 5 6 78 9 0 12 3 4 5 67 8 9 0 12 3 4 5 6 78 9 0 1 23 456 7890
str1 = '문자열 내부의 문자에 인덱스를 사용하여 접근하거나 슬라이싱 할 수 있다.'
str1
```

```python
# 인덱싱 - 한 개만 추출
print(str1[0])
type(str1[0])
```

```python
# 슬라이싱 - 특정 부분 추출
print(str1[0:5])   # 5번 인덱스 문자는 제외
print(str1[12:15])
print(str1[12:])
print(str1[-1:])
```

### 문자열 메소드

- 문자열 객체의 기능

```python
text = 'Hello, Python!'
# 대문자로 변환
print(text.upper())
# 소문자로 변환
print(text.lower())
```

```python
# 문자열의 등장 횟수
print(text.count('o'))
print(text.count('ll'))
```

```python
# 문자열의 첫번째 등장 위치(인덱스)
text = 'Hello, Python!'
print(text.find('o'))    # **
print(text.find('ll'))
print(text.find('z'))    # 찾는 문자열 없으면 -1
print(text.rfind('o'))   # rfind 메소드는 맨 뒤부터 찾음. **결과와 비교
print(text.find('o',6))   ### 인덱스 6번 부터 찾기
```

```python
# 문자열 양쪽 공백 제거
text='          Hello, Python!           '
print(text)
print(text.strip())
```

```python
'apple' == 'apple    '
```

```python
'apple' == 'apple    '.strip()
```

```python
### 문자열 분리
text = '김사과 반하나 오렌지 이메론'
names= text.split()   # 구분문자는 공백이 기본값
print(names)
type(names)
```

```python
text = '김사과, 반하나, 오렌지, 이메론'
names=text.split(', ')
print(names)
```

```python
# 문자열 결합
text = 'Hello'
print('🍕'.join('Hello'))
print(text.join('&&'))  # && 사이에 'Hello' 가 조인
print(text.join('&'))   # join 의 인자 문자열이 문자1개일 때는 join 못함.

# join 을 호출하는 text 문자열이 구분자 역할.
```

```python
# 문자열의 앞과 끝 검사하기
text = "Hello, Python!"
starts_with_hello = text.startswith("Hello")
ends_with_world = text.endswith("World!")

print(starts_with_hello)
print(ends_with_world)
```
