```python
# -5 ~ 256 까지의 정수 와 알파벳 문자열은 미리 메모리에 캐싱
# 항상 id 값이 동일합니다.
a=200     # a=257   값을 변경하면 id 결과 값에서 a, b주소가 다릅니다.
b=200     # b=257
print(a)
print(b)
```

```python
print(id(a))
print(id(b))
```

```python
a is b  # id(a) == id(b)
```

```python
a==b   # 저장된 값 비교
```

```python
# 불변 객체 : 값이 같으면 메모리에 하나만 저장
#  ㄴ  값을 바꿀 수 없다. 값이 바뀌면 메모리 주소가 바뀐다.
```

```python
name = 'twice'
print(name)
print(id(name))
```

```python
name = 'nujeans'    # 변수의 문자열이 다른 것이 할당. 변수의 주소가 바뀜
print(name)
print(id(name))
```

```python
message = 'twice'   # 변수명이 다르지만 문자열이 같으므로 주소가 같음.(자바의 불변객체)
print(message)
print(id(message))
```
