### 쉘(shell) 명령어 실행

- Google Colab은 구글 클라우드로 제공되는 가상환경. 리눅스 OS 기반
- Jupyter Notebook과 유사하게 구글 코랩에서 리눅스 쉘명령어를 실행할 수 있음.
- 쉘(Shell)은 컴퓨터 운영 체제에서 사용자와 시스템 사이의 중재자 역할을 하는 소프트웨어이며 리눅스는 ls, mkdir, rm, pwd 등의 명령어가 있음.
- 구글 코랩에서는 ! 기호와 함께 사용함.
